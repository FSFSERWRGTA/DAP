### 데이터프레임- 1) 딕셔너리

```python
#라이브러리 불러오기
import pandas as pd

#딕셔너리로로 데이터프레임 생성
dict_data = {'동물':['Dog','Cat','Tiger','Lion','Monkey'],'나이':[7,9,2,3,1]}
df1 = pd.DataFrame(dict_data)

#타입 확인
print(type(df1))

#데이터 확인
print(df1)
```

### 데이터프레임 - 2) 리스트

```python
#라이브러리 불러오기
import pandas as pd

#리스트로 데이터프레임 생성
df2 = pd.DataFrame([['Dog','7'], ['Cat','9'], ['Tiger','2'], ['Lion','3'],['Monkey','1']],
                  index = ['0','1','2','3','4'],
                  columns = ['동물','나이'])

#타입 확인
print(type(df2))
#데이터 확인
print(df2)
```

### 4) 조건을 이용하여 선택

```python
# 조건은 컬럼 데이터를 조건식과 비교하여 데이터를 추출하는 데이터베이스 SQL과 유사합니다
# 전체 데이터중 팁이 5 달러 이상
tips[tips.tip > 5]
```

```python
# and 조건 – 남자 손님이면서 비흡연자
tips[(tips['sex']=='Male') & (tips['smoker']=='No')]
```

### 5) 특정 조건 선택

```python
# 특정조건 선택시 ‘isin’ 함수를 많이 사용합니다.
# 일요일 이면 True
[tips['day'].isin(['Sun'])]
```

```python
# 팁을 1달러 지불한 고객
tips[tips['tip'].isin([1])]
```
